# Frame the problem and look at the big picture

**Supervised Learning Binary Classification Probelm:**

*This is a supervised learning classification problem as we have annotated / labelled dataset. In case we do not have labelled data available it is then it becomes Unsupervised learning problem.

Classification is applied when we have discreted(Fixed number of values) output variable, regression is applied when we have continous variable (infinite number of values).

Each data in the dataset has the expected output, set of classes are only two, hence it is a binary class classification problem(class: e= edible, p=posinious). If the number of classes are more than 2 then it is a multiclass classification problem.*

```python
#Function to build single bar graph

def count_plot_single(col, col_dict, c):
    
    #Set figure size
    f, ax = plt.subplots(figsize=(12,9))

    #Get the value count. value_counts() by default gives sorted result by count, however we need to sort based on index 
    #alphabetically as we are passing sorted values.
    counts = mushroom[col].value_counts().sort_index(ascending=True)
    pos = np.arange(len(counts))
    
    #Plot the bar Graph
    plt.bar(pos, counts, edgecolor='black', color=c)
    
    #Chart Formatting
    plt.xticks(pos, col_dict)
    plt.xlabel(f"{col.title()}", fontsize=16)
    plt.ylabel('Quantity of Mushrooms', fontsize=16)
    plt.title(f"Mushroom {col.title()} Quantity",fontsize=20)

    #Display the count on top of bar
    for i in pos:
            plt.text(i, counts[i], counts[i], ha = 'center', fontsize=14)
        
    plt.show()
```

**Visualizing the quantity of mushroom for different features in the dataset.**

```python
#Call function to display bar graph of mushroom quantities
col_dict=[{'edible':'e','poisonous':'p'},
          {"almond":"a","anise":"l","creosote":"c","fishy":"y",
             "foul":"f","musty":"m","none":"n","pungent":"p","spicy":"s"},
         {"orange":"b", "black":"c", "red":"e","gray":"g", "brown":"n","pink":"p","green":"r","purple":"u","yellow":"y", "blue":"w" }]
col=['class','odor','cap-color']
#Color assignment
c=['green','red','grey','blue','purple','pink','brown','orange','lightgreen','yellow']

for i in range(len(col)):
    count_plot_single(col[i],col_dict[i], c)
```

**Visualizing values for all the mushroom features w.r.t. label in order to find out which features are important.**

```python
#Function to build side by side bar chart w.r.t class label

def count_plot_side_by_side(col, col_dict):
    #Set figure size
    f, ax = plt.subplots(figsize=(12,9))
    
    pos = np.arange(len(col_dict))
    bar_width = 0.35
    class_dict=['Edible','Poisonous']
    
    #Chart Formatting
    plt.xticks(pos, col_dict)
    plt.xlabel(f"{col.title()}", fontsize=16)
    plt.ylabel('Quantity of Mushrooms', fontsize=16)
    plt.title(f"Mushroom {col.title()} Quantity",fontsize=20)
    
    #Get the data and divide into edible and non-edible
    newdf=mushroom.groupby([pd.Categorical(mushroom[col]),'class'])[col].count().unstack(fill_value=0)
    edible_count=newdf['e']
    poisonous_count=newdf['p']
    
    plt.bar(pos,edible_count,bar_width,color='green',edgecolor='black')
    plt.bar(pos+bar_width,poisonous_count,bar_width,color='red',edgecolor='black')
    
    plt.legend(col_dict,loc='best')
    
    #Display the count on top of bar
    for i in pos:
        plt.text(i, edible_count[i], edible_count[i], ha = 'center', fontsize=12)
        plt.text(i+bar_width, poisonous_count[i], poisonous_count[i], ha = 'center', fontsize=12)
        
    plt.show()
```

# Prepare the data for Machine Learning algorithms

## Converting categorical data to numerical.

*Machine Learning algorithms require numerical features. However, this dataset is composed of all categorical features.*

*There are different approaches of Encoding the data:*

**1) Label Encoding:**

*This is just assigning a number to each category, which we also call as factorize, similar to Ordinal Encoding however OrdinalEncoding works best is for ordinal variables. This technique works when the features are binary (only have 2 unique values) or when the features are ordinal categorical (that is, when the categories can be ranked). *

*In our case, only some of our features have 2 unique values and none of them are ordinal categorical, hence applied Label Encoding to those features with a binary set of values.**

```python
#For Feature selection correlation 
mush_feature = mushroom
le=LabelEncoder()
for column in mush_feature.columns:
    mush_feature[column] = le.fit_transform(mush_feature[column])
    if len(mushroom[column].value_counts()) == 2:
        le = LabelEncoder()
        le.fit(mushroom[column])
        mushroom[column] = le.transform(mushroom[column])
```

**2) One Hot Encoding**

*For the remaining features,we will use One Hot Encoding.*

*Essentially, this consists on creating a new binary feature representing each category. For instance, from the feature cap surface, which has 4 unique values (f, g, y and s), we create 4 binary features (cap_surface_f, cap_surface_g, cap_surface_y and cap_surface_s) indicating whether the category they represent was indeed that one or not. This means that, for any given instance (row), we will have exactly one of these 4 features equal to 1, and the other 3 equal to 0*

```python
#pd.get_dummies results to a Pandas DataFrame whereas OneHotEncoder results a SciPy CSR matrix.
mushroom = pd.get_dummies(mushroom)
mushroom.head()
```

## Linear Classifer Model - SGD Classifier

```python
from sklearn.linear_model import SGDClassifier

sgd_clf = SGDClassifier(random_state=42) #initialize mode
sgd_clf.fit(X_train,y_train) #fit model

print("SGDCLassifier Train Accuracy: {}%".format(round(sgd_clf.score(X_train, y_train)*100,2)))
print("SGDCLassifier Test Accuracy: {}%".format(round(sgd_clf.score(X_test, y_test)*100,2)))
```

```python
from sklearn.tree import DecisionTreeClassifier

tree_clf = DecisionTreeClassifier(max_depth=2, random_state=42)
tree_clf.fit(X_train, y_train)

print("Decision Tree Train Accuracy: {}%".format(round(tree_clf.score(X_train, y_train)*100,2)))
print("Decision Tree Test Accuracy: {}%".format(round(tree_clf.score(X_test, y_test)*100,2)))
```

## SVM Classification

```python
from sklearn.svm import SVC
svm = SVC(random_state=42, gamma="auto")
svm.fit(X_train, y_train)

print("SVM Train Accuracy: {}%".format(round(svm.score(X_train, y_train)*100,2)))
print("SVM Test Accuracy: {}%".format(round(svm.score(X_test, y_test)*100,2)))
```

## KNN Classification

```python
from sklearn.neighbors import KNeighborsClassifier
best_Kvalue = 0
best_score = 0
for i in range(1,10):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    if knn.score(X_test, y_test) > best_score:
        best_score = knn.score(X_train, y_train)
        best_Kvalue = i
        
print("Best KNN Value: {}".format(best_Kvalue))


print("KNN Train Accuracy: {}%".format(round(knn.score(X_train, y_train)*100,2)))
print("KNN Test Accuracy: {}%".format(round(knn.score(X_test, y_test)*100,2)))
```

## Logistic Regression Classification

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(solver="lbfgs", max_iter=500)
lr.fit(X_train, y_train)


print("Logistic Regression Train Accuracy: {}%".format(round(lr.score(X_train, y_train)*100,2)))
print("Logistic Regression Test Accuracy: {}%".format(round(lr.score(X_test, y_test)*100,2)))
```

## Confusion Matrix, Precision Score, Recall Score, F1Score

```python
def plot_confusion_matrix(y_test, y_test_pred, cm):
    x_axis_labels = ["Edible", "Poisonous"]
    y_axis_labels = ["Edible", "Poisonous"]
    f, ax = plt.subplots(figsize =(5,5))
    sns.heatmap(cm, annot = True, linewidths=0.2, linecolor="black", fmt = ".0f", ax=ax, cmap="Blues", xticklabels=x_axis_labels, yticklabels=y_axis_labels)
    plt.xlabel("PREDICTED LABEL")
    plt.ylabel("TRUE LABEL")
    plt.show()
```

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix, recall_score, precision_score, f1_score

def confusion_matrix_model(Model):
    y_test_pred = cross_val_predict(Model,X_test, y_test, cv=3)
    cm = confusion_matrix(y_test,y_test_pred)
    
    #Precision Score
    ps = precision_score(y_test,y_test_pred)
    
    #Recall Score
    rs= recall_score(y_test,y_test_pred)
    
    #F1 Score - Instead of computing precision and recall everytime, 
    #we prefer a single mmetric which combines both precision and recall, This single metric is F1 score.
    # This is harmonic mean of precision and recall
    # F1=2/(1/precision) + (1/recall)
    f1sc = f1_score(y_test,y_test_pred)
    print("\n" + "Confusion Matrix")
    plot_confusion_matrix(y_test, y_test_pred, cm)
    print("\n" + "Precision Score")
    print(ps)
    print("\n" + "Recall Score")
    print(rs)
    print("\n" +  "F1 Score")
    print(f1sc)
```

```python
#SCD
print('\033[1m' + "SGC Classifier: " + '\033[1m')
confusion_matrix_model(sgd_clf)

#LR
print("\n"  + '\033[1m' + "Logistic Regression" + '\033[1m' )
confusion_matrix_model(lr)

#SVM
print('\033[1m' + "\n" + "SVM" + '\033[1m')
confusion_matrix_model(svm)

#KNN
print('\033[1m'+ "\n" + "KNN" + '\033[1m')
confusion_matrix_model(tree_clf)

#Decision Tree
print('\033[1m' + "\n" + "Decision Tree" + '\033[1m')
confusion_matrix_model(tree_clf)
```
